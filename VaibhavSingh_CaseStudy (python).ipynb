```python
shopping_cart={'S.no1':{'item':'Biscuits-a pack of 20','quantity':310,'cost':2.50},
               'S.no2':{'item':'Cereals-2 pounds/pack','quantity':260,'cost':3.50},
               'S.no3':{'item':'Chicken-1 pounds/pack','quantity':290,'cost':13.30},
               'S.no4':{'item':'Oats-4 pounds/pack','quantity':340,'cost':3.90},
               'S.no5':{'item':'Rice-2 pounds/pack','quantity':470,'cost':4.50}}
```

```python
#creating the inventory of items in a tabular form that are available in shopping_cart.
def inventory():
    print('%46s'%('ESSENTIAL ITEMS AVAILABLE'))
    print('%47s'%('*'*27))
    print('%9s %20s %20s %10s'%('S.NO','ITEMS','QUANTITY','COST'))
    print('%64s'%('`'*60))
    for Sno_key,details in shopping_cart.items():
        Sno_item=details['item']
        Sno_quantity=details['quantity']
        Sno_cost=details['cost']
        print('%10s %26s %10d %13.2f$'%(Sno_key,Sno_item,Sno_quantity,Sno_cost))
    print('%64s'%('`'*60))
    
# customer name and delivery details.

def cus_details():
    print('')
    print('Fill in your delivery details:- ')
    print('')
    name=input('Enter your name: ')
    add=input('Enter your address: ')

# calculating delivery charges base on the distance.

def delivery_charges():
    global distance
    global delivery_charge
    print('')
    distance=int(input('Distance from store in miles: '))
    delivery_charge=0
    if distance>0 and distance<=5:
        delivery_charge+=10
    elif distance>5 and distance<=10:
        delivery_charge+=12
    elif distance>10 and distance<=15:
        delivery_charge+=14
    elif distance>15:
        delivery_charge+=20
    else:
        delivery_charge+=0
    return delivery_charge
    
# final bill.

def final_bill(requirement):
    total_cost=0
    for item, quantity in requirement.items():
        total_cost+=shopping_cart[item]['cost']*quantity
    return total_cost

# update stock for the next user.

def update_stock(requirement):
    for item,quantity in requirement.items():
        shopping_cart[item]['quantity']-=quantity
        
# Asking the customer to enter the number of quantities required for each item.
       
while True:
    inventory()   # displaying the main inventory of available items.
    
    requirement={}    # creating a temp. empty dictonary to store the quantity of required items.
    
    for item,details in shopping_cart.items():
        inner_item=details['item']
        quantity_required=int(input(f'how many packets of "{inner_item}" are required? '))
        if quantity_required>-1:
            if quantity_required<=shopping_cart['S.no1']['quantity']:
                requirement[item]=quantity_required
            else:
                print('*'*5,'item is out of stock','*'*5)
        else:
            print('Invalid entry')
    total_cost=final_bill(requirement)  # addressing the final_bill function with a variable name.
    delivery_charge=delivery_charges()  # addressing the delivery_charges function with a variable name.
    if requirement:
        cus_details()  # calling customer details function.
        
        # Display the final Bill in a tabular format with total cost.
        zipped_dict=zip(shopping_cart.items(),requirement.items())
        print('')
        print('%36s'%('BILL'))
        print('%37s'%('*'*6))
        print('%9s %20s %19s %15s'%('S.NO','ITEMS','QUANTITY','TOTAL COST'))
        print('%67s'%('`'*64))
        for (key1,value1),(key2,value2) in zipped_dict:
            sno_item=value1['item']
            sno_cost=value1['cost']*value2
            print('%10s %26s %9d %15.2f$'%(key1,sno_item,value2,sno_cost))
        print('%67s'%('`'*64))
        
        # total cost with delivery charges.
        
        print(f'total item cost: ${total_cost}')
        total_bill_amount=total_cost+delivery_charge
        print(f'total bill amount = total cost + delivery charge = ${total_bill_amount}')
        
        update_stock(requirement) # calling stock update function.
    else:
        print('No item selected')
    con_shop=input('Do you want to add more items? y/n: ')
    
    # program should stop running if the user press 'n' but continue the loop if 'y' is pressed.
    if con_shop=='n':
        print('%s'%('*'*45),'THANKYOU FOR SHOPPING WITH US','%s'%('*'*51))
        break
```
